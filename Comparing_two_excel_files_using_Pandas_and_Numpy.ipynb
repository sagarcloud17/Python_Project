```python
#importing the required libraries
import pandas as pd
import numpy as np
```

```python
#loading the table1
data1=pd.read_excel("/sample1.xlsx")
data1
```

```python
#loading the table2
data2=pd.read_excel("/Sample2.xlsx")
data2
```

```python
#compare the rows and cloumns in table1
data1.shape
```

```python
#compare the rows and cloumns in table2
data2.shape
```

```python
#comparing the row and columns in both the tables
comparevalues=data1==data2
```

```python
#to display the differences cloumn wise
diff = data1.compare(data2, align_axis=1)
diff.to_excel("/output_sample.xlsx")
diff
#self=table1, other=table2
```

```python
#to display the differences row wise
diff = data1.compare(data2, align_axis=0)
diff.to_excel("/output_sample2.xlsx")
diff
#self=sampledata1, other=sampledata2
```

```python
#to display the entire data with differnece
diff = data1.compare(data2, keep_shape=True, keep_equal=True)
diff
#self=sampledata1, #other=sampledata2
```
